```python
import os
import mne
import pickle
```

```python
# Path to your EEG files directory
data_dir = '/home/donaf-strange/LAB_WORK/eeg_arithmetic_project/data/files'

# Dictionary to store EEG data
task_eeg_dict = {}

# Loop through all files in the directory
for fname in os.listdir(data_dir):
    # Only process task files (_2.edf)
    if fname.endswith('_1.edf') and fname.startswith('Subject'):
        # Extract subject number (e.g., Subject07_2.edf -> '07')
        subject_no = fname.split('_')[0].replace('Subject', '').zfill(2)
        file_path = os.path.join(data_dir, fname)
        # Load EEG data using MNE
        raw = mne.io.read_raw_edf(file_path, preload=True, verbose=False)
        # Store in dictionary
        task_eeg_dict[subject_no] = raw

print(f"Loaded {len(task_eeg_dict)} task EEG files.")

# Save the dictionary to a file
save_path = '/home/donaf-strange/LAB_WORK/eeg_arithmetic_project/data/all_rest_eeg.pkl'
with open(save_path, 'wb') as f:
    pickle.dump(task_eeg_dict, f)
print(f"Task EEG data saved to {save_path}")

# Example: How to load the saved file later
# with open(save_path, 'rb') as f:
#     task_eeg_dict = pickle.load(f)
```

```python
import os
import mne
import pickle

data_dir = '/home/donaf-strange/LAB_WORK/eeg_arithmetic_project/data/files'
task_eeg_dict = {}

for fname in os.listdir(data_dir):
    if fname.endswith('_1.edf') and fname.startswith('Subject'):
        file_path = os.path.join(data_dir, fname)
        raw = mne.io.read_raw_edf(file_path, preload=True, verbose=False)
        participant_no = None
        try:
            if 'subject_info' in raw.info and raw.info['subject_info'] is not None:
                if 'his_id' in raw.info['subject_info']:
                    participant_no = str(raw.info['subject_info']['his_id'])
                elif 'id' in raw.info['subject_info']:
                    participant_no = str(raw.info['subject_info']['id'])
            if not participant_no:
                participant_no = str(raw.info.get('subject_info', {}).get('name', None))
            if not participant_no and 'experimenter' in raw.info:
                participant_no = str(raw.info['experimenter'])
        except Exception as e:
            print(f"Could not extract participant number from {fname}: {e}")
        # Only skip if participant_no is truly None or empty (not if it's '0')
        if participant_no is None or participant_no.strip() == '':
            print(f"Warning: Could not find participant number for {fname}. Skipping.")
            continue
        participant_no = participant_no.strip()  # Do not lstrip('0')
        task_eeg_dict[participant_no] = raw

print(f"Loaded {len(task_eeg_dict)} task EEG files (by participant number).")

save_path = '/home/donaf-strange/LAB_WORK/eeg_arithmetic_project/data/all_rest_eeg_by_participant.pkl'
with open(save_path, 'wb') as f:
    pickle.dump(task_eeg_dict, f)
print(f"Task EEG data saved to {save_path}")
```
